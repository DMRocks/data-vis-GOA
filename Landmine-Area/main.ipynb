```python
#Import our tools and Data
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

url = 'https://raw.githubusercontent.com/DMRocks/Data-Vis-Landmine-Area/main/Areas%20contaminated%20by%20landmine%20(sq.km).cvs'
data = pd.read_csv(url, index_col = 'Year')

data.head()
```

```python
data.plot.area()
```

```python
columns_to_use = ['Year', 'Afghanistan', 'Bosnia and Herzegovina', 'Cambodia', 'Croatia', 'Egypt, Arab Rep.', 'World']

chart = pd.read_csv(url, index_col= 'Year')

chart.plot(kind = 'bar', stacked = True)
```

```python
#Plotly

columns_to_use = ['Afghanistan', 'Bosnia and Herzegovina', 'Cambodia', 'Croatia', 'Egypt, Arab Rep.', 'World']

print(data.index)

print(columns_to_use)

bar_chart = px.bar(data, x = data.index, y = columns_to_use)


bar_chart.update_layout(
    title = 'Top Countries Contaminated by Landmines World Bank',
    yaxis_title = 'Square Kilometers',
    legend_title_text = 'Country',
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 1
    ),
    yaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 3000
    )
)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots


fig = make_subplots(rows= 1, cols= 2, specs=[[{'type':'domain'}, {'type':'domain'}]])

fig.add_trace(go.Pie(labels= columns_to_use, values= data.iloc[0, 1:], name= 'Landmine Area in 1999'), 1 , 1)

fig.add_trace(go.Pie(labels= columns_to_use, values= data.iloc[8, 1:], name= 'Landmine Area in 2007'), 1 , 2)

fig.update_traces(hole= .4)

fig.update_layout (
    title_text= 'Top Countries Contaminated by Landmines World Bank',
    annotations= [ 
        dict(text= '1999', x= 0.165, y= 0.5, font_size=20, showarrow= False),
        dict(text= '2007', x= 0.83, y= 0.5, font_size=20, showarrow= False)
    ]


)
```
