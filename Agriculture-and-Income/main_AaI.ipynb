```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import statsmodels


url = "https://raw.githubusercontent.com/DMRocks/Agriculture-v-Personal-Income/main/Agriculture%20Based%20Countries%20Effect%20on%20Personal%20Income%20-%20Combined%20Vaules%20Only.csv"

data = pd.read_csv(url, index_col = "country")

#This block of code will do two things, check if countrys have three vaules, and it will format from the github into something useable

dups_color = data.pivot_table(columns=['country'], aggfunc='size')

dups_color = dups_color.to_dict()

dups_color = {key:val for key, val in dups_color.items() if val != 2}

coloumns_to_use = list(dups_color.keys())

data_useable = data.loc[coloumns_to_use]

data_useable = data_useable.drop(index=['Holy See', 'Liechtenstein'])

print(data_useable)
```

```python
#Oh my lord this part took my forever! However, I am done. This part of the code will format it like it needs to be formatted.

temp_list = list(data_useable.index)

countrys_to_loop = []

for x in temp_list:
    if x not in countrys_to_loop:
        countrys_to_loop.append(x)

df = pd.DataFrame(columns = ['Country', 'Year', 'Population', 'Agriculture', 'Income'])

primer = 0

for country in countrys_to_loop:
    
    year = 1959

    primer_y = 1

    while year != 2019:

        primer_x = primer

        population_vaule = data_useable.iloc[primer_x, primer_y]
        primer_x += 1 
        
        income_vaule = data_useable.iloc[primer_x, primer_y]
        primer_x += 1 

        agriculture_vaule = data_useable.iloc[primer_x, primer_y] 
        primer_x += 1 

        temp_list = [country, year, population_vaule, income_vaule, agriculture_vaule]

        df_length = len(df)
        df.loc[df_length] = temp_list
        temp_list = []

        year += 1

        primer_y += 1

    primer +=3

print(df)

```

```python
#Quick Python Scatter Plot

fig = px.scatter(df.query("Year==2018"), x = 'Income', y = 'Agriculture', size= 'Population', color= 'Country', size_max=30)

fig.show()
```

```python
#Python Slider

fig = px.scatter(df, x = 'Income', y = 'Agriculture', size= 'Population', animation_frame="Year", animation_group="Country",
     color="Country", hover_name="Country", 
     size_max=60, range_x=[0, 47000], 
     range_y=[0, df['Agriculture'].max()])

fig.update_layout(
    legend=dict(
        bgcolor="White",
        bordercolor="LightSteelBlue",
        borderwidth= 2,
        itemclick = 'toggleothers',
    ),
     title = '<b>Agriculture Based Economys Against Personal Income (1959-2018)</b>'+'<br>'+'<span style="font-size: 12px;">Data source: Gapminder. Size indicates population size.',
     yaxis_title = 'Agriculture percent of GDP',
     xaxis_title = 'Income Per Person (GDP per Capita)'
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import statsmodels
import numpy as np


url = "https://raw.githubusercontent.com/DMRocks/Agriculture-v-Personal-Income/main/final_formated.csv"

df = pd.read_csv(url)


#Python Slider

fig = px.scatter(df, x = 'Income', y = 'Agriculture', size= 'Population', animation_frame="Year", animation_group="Country",
     color="Country", hover_name="Country", 
     size_max=60, range_x=[0, 47000], 
     range_y=[0, df['Agriculture'].max()])

fig.update_layout(
    legend=dict(
        bgcolor="White",
        bordercolor="LightSteelBlue",
        borderwidth= 2,
        itemclick = 'toggleothers',
    ),
     title = '<b> Agriculture Based Economys Against Personal Income (1959-2018)</b>'+'<br>'+'<span style="font-size: 12px;">Data source: Gapminder. Size indicates population size.',
     yaxis_title = 'Agriculture percent of GDP',
     xaxis_title = 'Income Per Person (GDP per Capita)'
)

fig.add_shape(type="rect",
    x0=7000, y0=25, x1= 50000, y1= 100,
    line=dict(
        color="Red",
        dash='dash',
        width=2,
    ),
)

fig.add_annotation(text = '<b> For most countries, to past 7k of personal income, </b>',
                        showarrow = False,
                        font = dict(size = 17),
                        x=26000, y=60)

fig.add_annotation(text = '<b> Agriculture must make up lower than 25% of their GDP. </b>',
                        showarrow = False,
                        font = dict(size = 17),
                        x=28000, y=50)


fig.write_html('Agriculture.html')
```
